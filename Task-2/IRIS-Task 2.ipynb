# Step 3:- Finding the optimum number of clustering using k-means clustering algorithm(This is our main objective)

```python
x=iris_df.iloc[:,[0,1,2,3]].values

wcss=[]
for i in range(1,6):
    km=KMeans(init="k-means++", n_clusters=i, n_init=10, max_iter=300, random_state=0)
    
    km.fit(x)
    wcss.append(km.inertia_)
    print('k:', i, "wcss:", km.inertia_)
```

```python
#Plotting the result into line graph
#The ELBOW method



plt.plot(3, 78.85144142614601, color='black', marker='o', markersize=10)
plt.plot(range(1,6),wcss,marker='o')

plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within cluster sum of squares')
plt.show()
```

```python
wcss
```

# Step 4:- Initializing K-means with the optimum number of cluster and predicting the values

```python
#Fitting the k-means in data
km=KMeans(init="k-means++",n_clusters=3, n_init=10, max_iter=300,random_state=0)

y_km=km.fit_predict(x)
```
